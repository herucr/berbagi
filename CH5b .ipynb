```python
# import library yang diperlukan
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
N =40
p = 0.1
G = nx.gnp_random_graph(N,p) # Gilbert random graph
nx.draw(G, pos=nx.circular_layout(G))
print("Aktual banyak edge             = " + str(G.number_of_edges()))
```

```python
L = 80
G = nx.gnm_random_graph(N,L) # Erdos-Renyi random graph
nx.draw(G, pos=nx.circular_layout(G))
print("Aktual banyak edge             = " + str(G.number_of_edges()))
```

```python
N=15
k=4
p=0.1
G = nx.watts_strogatz_graph(N,k,p) # smallworld model network : Watts and Strogatz
nx.draw(G, pos=nx.circular_layout(G))
```

```python
sequence = nx.random_powerlaw_tree_sequence(20, tries=5000)
#G = nx.configuration_model(D) # network with degree sequence D
G = nx.configuration_model(sequence)
sequence
nx.draw(G,pos=nx.circular_layout(G))
```

```python
#pemilihan node tanpa preferential attachment
nodes = [1, 2, 3, 4]
selected_node = random.choice(nodes)
selected_node
```

```python
#pemilihan node dengan  preferential attachment
nodes = [1, 2, 3, 4]
degrees = [3, 1, 2, 2]
selected_node = random.choices(nodes, degrees)
selected_node
```
